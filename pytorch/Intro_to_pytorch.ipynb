# PyTorch

### What Deep learning is good for ? 
1. Complex problem with large number of rules
2. Large Dataset 
3. To generalize well for unseen data

### What deep Learning is not typically good for?
1. It contains a black box where the parameters are determined by the model itself where explanation is not provided
2. Problem is simple and can be easily solved using traditional method
3. When errors are unacceptable : Outputs of deep learning models are probabilistic
4. When we do not have large amount of data

**Machine Learning** - more suitable for strucutred data : data in the form of rows and columns
**Deep Learning** - more suitable for unstrucutred data : Data with no shape , eg : voice , video , text

### Why PyTorch? 
- Most popular Deep Learning framework
- fast development of code in python and allows to run on GPUs
- Able to access other pre-trained models  and perform transfer learning
- allows easy pre-processing of data , model development and deployment
